```python
# Import scipy for statistical functions
import scipy.stats as stats

# Load the combined dataset
# Load dataset using pandas
ecom = pd.read_csv('e-commerce.csv')

# Function to analyze a column
def analyze_column(df, col_name):
    description = {}
    description['Column Name'] = col_name

    # Number of empty rows
    description['Empty Rows'] = df[col_name].isnull().sum()

    # Data Type
    description['Data Type'] = df[col_name].dtype

    # Range (min, max) for numeric columns
    if pd.api.types.is_numeric_dtype(df[col_name]):
        description['Min'] = df[col_name].min()
        description['Max'] = df[col_name].max()
        description['Mean'] = df[col_name].mean()
        description['Median'] = df[col_name].median()

        # Mode (handling single and multiple modes)
        mode_result = stats.mode(df[col_name], keepdims=True)  # Ensure it returns an array
        description['Mode'] = mode_result.mode[0] if mode_result.mode.size > 0 else None

        if mode_result.mode.size > 1:
            description['Other Modes'] = list(mode_result.mode[1:])

    # Unique values for non-numeric columns
    elif pd.api.types.is_object_dtype(df[col_name]):
        description['Unique Values'] = df[col_name].unique()

    return description

# Analyze each column
column_analyses = [analyze_column(ecom, col) for col in ecom.columns]

# Create a DataFrame from the analysis results
analysis_df = pd.DataFrame(column_analyses)
analysis_df
```

```python
def fill_missing_with_iqr_random(df, seed=42):
    """
    Fill missing values in numeric columns with random values
    between the 25th and 75th percentiles (IQR), only at missing positions.

    Parameters:
        df (pd.DataFrame): Input DataFrame.
        seed (int): Random seed for reproducibility.

    Returns:
        pd.DataFrame: DataFrame with filled missing numeric values.
    """
    np.random.seed(seed)
    df_filled = df.copy()

    for col in df_filled.select_dtypes(include=[np.number]).columns:
        q1 = df_filled[col].quantile(0.25)
        q3 = df_filled[col].quantile(0.75)
        missing_mask = df_filled[col].isna()

        if missing_mask.any():
            df_filled.loc[missing_mask, col] = np.random.uniform(q1, q3, size=missing_mask.sum())

    return df_filled
```

```python
# Import pandas for data manipulation and analysis
import pandas as pd
# Import numpy for numerical operations
import numpy as np

# Function to clean a date column with corrupted string entries
def clean_date_column(df, column_name):
    # Step 1: Convert to datetime, coercing invalid values to NaT
    df[column_name] = pd.to_datetime(df[column_name], errors='coerce')

    # Step 2: Drop NaTs for percentile calculation
    valid_dates = df[column_name].dropna()

    if len(valid_dates) == 0:
        print(f"No valid dates found in '{column_name}' to generate replacements.")
        return df

    # Step 3: Calculate percentiles
    p25 = valid_dates.quantile(0.25)
    p75 = valid_dates.quantile(0.75)

    # Step 4: Generate random dates and fill NaT
    mask = df[column_name].isna()
    random_dates = pd.to_datetime(
        np.random.uniform(p25.value, p75.value, size=mask.sum())
    ).floor('D')

    df.loc[mask, column_name] = random_dates
    print(f"{mask.sum()} invalid entries in '{column_name}' replaced with random dates.")

    return df

# Apply to all target columns
ecom = clean_date_column(ecom, 'Date')
ecom = clean_date_column(ecom, 'DATE')
ecom = clean_date_column(ecom, 'Months')
```

```python
# Import pandas for data manipulation and analysis
import pandas as pd
# Import numpy for numerical operations
import numpy as np
# Import train_test_split for splitting data into train/test sets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb
# Import MSE for regression evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming ecom is already loaded
# Load dataset using pandas
ecom = pd.read_csv('train&testdata.csv')  # Load if not already loaded

def preprocess_data(df):
    print("Starting preprocessing...", flush=True)
    if 'Amount' not in df.columns:
        raise KeyError("Target column 'Amount' not found.")

    y = df['Amount']
    X = df.copy()

    categorical_cols = X.select_dtypes(include=['object', 'bool']).columns.tolist()
    print(f"Categorical columns: {categorical_cols}", flush=True)

    label_encoders = {}
    for col in categorical_cols:
        print(f"Encoding: {col}", flush=True)
        X[col] = X[col].astype(str)
        X[col] = X[col].fillna('NA')
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col])
        label_encoders[col] = le

    numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
    for col in numeric_cols:
        X[col] = X[col].fillna(X[col].median())

    print("Preprocessing complete.", flush=True)
    return X, y, label_encoders

def train_and_evaluate(X, y):
    print("Splitting data and training model...", flush=True)
# Import train_test_split for splitting data into train/test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    model = xgb.XGBRegressor(objective='reg:squarederror',
                             n_estimators=200,
                             learning_rate=0.1,
                             random_state=42)

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
# Import MSE for regression evaluation
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print("Model evaluation metrics:", flush=True)
    print(f"MAE: {mae:.2f}", flush=True)
    print(f"MSE: {mse:.2f}", flush=True)
    print(f"RÂ²: {r2:.3f}", flush=True)

    return model

# Run the workflow
X, y, label_encoders = preprocess_data(ecom)
model = train_and_evaluate(X, y)
```
